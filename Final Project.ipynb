```python
address = 'San Jos, Costa Rica'

geolocator = Nominatim(user_agent="molina")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

```python
search_query='Vete'
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, RADIUS, LIMIT)
url
```

```python
results = requests.get(url).json()
results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
dataframe.shape
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name','location.distance', 'location.lat','location.lng', 'location.formattedAddress' ]  + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]
```

```python
dataframe_filtered.head()
```

```python
veterinarias= dataframe_filtered.drop(['name','location.formattedAddress','id', 'location.distance'], 1)
```

```python
veterinarias2=veterinarias.drop(columns='Cluster Labels')
```

```python
veterinarias2.head()
```

```python
#Clustering in 3 clusters

# set number of clusters
kclusters = 3

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(veterinarias2)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
dataframe_filtered.insert(0, 'Cluster Labels2', kmeans.labels_)
```

```python
dataframe_filtered.head()
```

```python
final=dataframe_filtered.drop(columns='Cluster Labels')
```

```python
final.head()
```

```python
#Visualization
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(final['location.lat'], final['location.lng'], final['name'], final['Cluster Labels2']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
